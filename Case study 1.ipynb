### 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
#finding num of buyers participated
a=df['Buyer Gender'].value_counts()
a
```

```python
#plotting bar graph

#1

x = list(a.index)
y = list(a)
fig, ax = plt.subplots()  
# the width of the bars
width = 0.75 
# the x locations for the groups
ind = np.arange(len(y))  
ax.barh(ind, y, width, color="yellow")
ax.set_yticks(ind+width/2)
ax.set_yticklabels(x, minor=False)
#add value labels into bar
for i, v in enumerate(y):
    ax.text(v + .25, i + .25, str(v), color='green', fontweight='bold') 
plt.title('Male vs Female')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 4.Find the top 5 cars based on their sales price. 

```python
 # values of top 5 sales price
df['Sale Price']. nlargest(n=5) 
```
